# The Best Neighborhood

Based on our three metrics - signalized intersections, speeding compliance, and traffic safety signs - we can now determine which Pittsburgh neighborhood has the best safety around motor vehicles.

```python
import pandas as pd
import matplotlib.pyplot as plt

signalInter = pd.read_csv("SignalizedIntersections.csv")

#count the amount of signalized intersections per neighborhood

neighborhoodDict = dict()

for value, row in signalInter.iterrows():
    neighborhood = row["neighborhood"]
    if neighborhood in neighborhoodDict and pd.notna(neighborhood):
        neighborhoodDict[neighborhood] += 1
    else:
        neighborhoodDict[neighborhood] = 1

dataFrameSums = pd.Series(neighborhoodDict)

dataFrameSums= {'total_signals':dataFrameSums}
dataFrameSums=pd.DataFrame(dataFrameSums)

reduced = dataFrameSums[dataFrameSums["total_signals"]>10]
increasingSorted = reduced.sort_values(by = 'total_signals', ascending = False)
#plot a bar graph

increasingSorted.plot.bar(title = 'Total Signals per Neighborhood', figsize=(10, 10))
```

```python
trafficData = pd.read_csv("SpeedingData.csv", index_col="_id")
df = trafficData[["neighborhood", "percent_over_limit"]]
df = df.dropna(subset=["percent_over_limit"])
avg = df.groupby("neighborhood")["percent_over_limit"].mean()
avg = avg.nsmallest(n=20)
avg.plot(x='Neighborhoods', y='Percent over Speed Limit', kind='bar', title='Percent over Speed Limit in Each Neighborhood', color='#289c3e')
```

## Why Central Business District is the Best Neighborhood

After analyzing three comprehensive metrics across all Pittsburgh neighborhoods - signalized intersections, speeding compliance, and stop sign infrastructure - our data clearly identifies Central Business District as the safest neighborhood for motor vehicle safety.

Central Business District excels across all three dimensions with 87 signalized intersections (the most in Pittsburgh), only 12.3% of drivers exceeding speed limits (well below the 28.4% city average), and 3,491 stop signs providing comprehensive intersection control. What makes this finding significant isn't just high scores in individual categories, but the combination of all three working together. The extensive infrastructure demonstrates maximum city investment in traffic safety, the low speeding rates prove this infrastructure actually changes driver behavior, and the comprehensive stop sign coverage creates layered protection from major intersections down to residential streets.

This combination translates to tangible safety benefits: controlled intersections reduce accident risk for drivers, pedestrians can navigate confidently knowing every crossing is regulated, and the systematic traffic management enables safe integration of buses, bikes, and emergency vehicles. While Central Business District's scores reflect its unique role as Pittsburgh's urban core - requiring maximum regulation to safely accommodate thousands of daily commuters and visitors - the data unambiguously shows this is where the city has achieved its highest level of vehicular safety infrastructure and compliance.
