```python
import pandas as pd
import numpy as np
trafficData = pd.read_csv("Traffic.csv", index_col="_id")
df = trafficData[["neighborhood", "percent_over_limit"]]
df = df.dropna(subset=["percent_over_limit"])
avg = df.groupby("neighborhood")["percent_over_limit"].mean()
avg = avg.nsmallest(n=20)
avg.plot(x='Neighborhoods', y='Percent over Speed Limit', kind='bar', title='Percent over Speed Limit in Each Neighborhood', color='#289c3e')
```

This graph was created by taking the given daily percent of drivers going over the speed limit in each neighborhood and finding the average based on each location. This graph shows the 20 lowest percentages, labeled by neighborhood. The lower this number is, the safer drivers are being and the safer pedestrians are around motor vehicles. 

```python
trafficData = pd.read_csv("Traffic.csv", index_col="_id")
df = trafficData[["neighborhood", "speed95_percent"]]
df = df.dropna(subset=["speed95_percent"])
avg = df.groupby("neighborhood")["speed95_percent"].mean()
avg = avg.nsmallest(n=20)
avg.plot(x='Neighborhoods', y='95th Percentile of Speed', kind='bar', title='95th Percentile of Speed in Each Neighborhood', color = '#9548e8')
```

This graph takes the daily 95th percentile speed at each neighborhood and averages them out. This shows the lowest 20 neighborhoods in this data. The higher this number is, the more reckless people are driving and the more dangerous that is for pedestrians, other drivers and cyclists.

```python
trafficData = pd.read_csv("Traffic.csv", index_col="_id")
df = trafficData[["neighborhood", "max_speed"]]
df = df.dropna(subset=["max_speed"])
avg = df.groupby("neighborhood")["max_speed"].mean()
avg = avg.nsmallest(n=20)
avg.plot(x='Neighborhoods', y='Max Speed', kind='bar', title='Max Speed in Each Neighborhood', color='#ffe438')
```

This graph shows the average daily max speed in each neighborhood. It shows the 20 neighborhoods with the lowest max speed, the lower this number is the more safe drivers are being around this area.
