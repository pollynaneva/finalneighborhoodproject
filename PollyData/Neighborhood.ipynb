```python
import pandas as pd
import numpy as np
trafficData = pd.read_csv("Traffic.csv", index_col="_id")
df = trafficData[["neighborhood", "percent_over_limit"]]
df = df.dropna(subset=["percent_over_limit"])
avg = df.groupby("neighborhood")["percent_over_limit"].mean()
avg = avg.nsmallest(n=20)
avg.plot(x='Neighborhoods', y='Percent over Speed Limit', kind='bar', title='Percent over Speed Limit in Each Neighborhood')
```
