In this dataset, we will explore data on the Traffic and safety signs in each neighborhood. I am analyzing this by counting and analyzing traffic control and safety signs. My hypothesis is that neighborhoods with more safety-oriented signage are "better" because they have better-controlled traffic flow, prioritize pedestrian safety, are more family-friendly, and show city investment in resident safety.

stop signs: Indicate controlled intersections and traffic calming. More stop signs = slower, safer streets.
pedestrian safety signs: Include "Watch Children", "Pedestrian", and "School Speed Limit" signs. These indicate family-friendly neighborhoods where pedestrian safety is prioritized.

More traffic safety infrastructure indicates:

- **Safer streets** - Controlled intersections reduce accidents and speeding
- **Pedestrian-friendly** - Safe for walking, biking, and children playing outside
- **Family-oriented** - Child safety signs show residential neighborhoods with kids
- **City investment** - More signs = more maintenance and attention from Pittsburgh
- **Complements teammate's data** - My teammate analyzes speeding violations; I analyze the infrastructure designed to prevent those problems

Together with my group's other metrics, this helps identify neighborhoods that are both safe on paper (infrastructure) and safe in practice (low violations).

```python
import pandas as pd 
import numpy as np

signData = pd.read_csv("trafficSigns.csv", index_col="_id")

# Count pedestrian-related signs
pedestrian_types = ['Watch Children', 'Pedestrian', 'School Speed Limit']
df = signData[["neighborhood", "description"]]
df = df[df["description"].isin(pedestrian_types)]
df = df.dropna(subset=["neighborhood"])
pedCount = df.groupby("neighborhood").size()
pedCount = pedCount.nlargest(n=20)
pedCount.plot(x='Neighborhoods', y='Number of Pedestrian Safety Signs', kind='bar', 
              title='Pedestrian Safety Signs in Each Neighborhood', color='green')
```

The amount of pedestrian safety signs indicate family-friendly neighborhoods where speeding is especially dangerous. These neighborhoods prioritize resident safety over fast traffic flow.

```python
import pandas as pd 
import numpy as np

signData = pd.read_csv("trafficSigns.csv", index_col="_id")

safety_types = ['Stop', 'Speed Limit', 'Watch Children', 'Pedestrian', 'School Speed Limit', 'Yield', 'Do Not Enter', 'No Turn on Red']
signData['is_traffic_safety'] = signData['description'].isin(safety_types)

df = signData[["neighborhood", "is_traffic_safety"]]
df = df.dropna(subset=["neighborhood"])
safetyScore = df.groupby("neighborhood")["is_traffic_safety"].sum()
safetyScore = safetyScore.nlargest(n=20)
safetyScore.plot(x='Neighborhoods', y='Traffic Safety Infrastructure Score', kind='bar', 
                 title='Total Traffic Safety Infrastructure by Neighborhood', color='blue')
```

This composite score combines all traffic control and safety signs. Neighborhoods with higher scores have better infrastructure to prevent speeding problems.
